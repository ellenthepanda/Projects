```python
import csv

data= []
rocks = []
headers = ['x_pos', 'y_pos', 'weight', 'weight_info']

with open('/Users/ellenmuller/Downloads/rocks.txt', 'r') as txt_file:

    for row in txt_file:
        data.append(row.split("  "))

print(data)

for rock in data:
    rock_data={}
    rock_data['x_pos']=rock[0]
    rock_data['y_pos']=rock[1]
    rock_data['weight']=rock[2].replace("\n", "")
    
    if len(rock)>3: 
        rock_data["weight_info"]=rock[3].replace("\n", "")   
    else:
        rock_data["weight_info"]=None
    
    rocks.append(rock_data)
    print(rock_data)
 
print(rocks) 

with open('rocks.csv', 'w') as file:
    writer = csv.DictWriter(file, headers)
    writer.writeheader()
    for rows in rocks:
        writer.writerow(rows)
        
```
